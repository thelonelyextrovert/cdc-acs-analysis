### Data Processing

Since the `Data_Value`, which shows us the percentage of individuals affected by a particular health condition, is in percentage format, we need to obtain the actual number of people affected for our analysis. To do this, we create two new columns where values are stored as integers:

1. **TotalPatients**: This column calculates the total number of individuals affected by the health condition across the entire population. 

2. **TotalPatients18plus**: This column calculates the number of individuals aged 18 and over affected by the health condition. The formula used is:

These new columns will allow us to work with counts of people, making our analysis easier to understand and enabling better comparisons across different locations and group types.

```python
# Calculate the total number of patients based on the Data_Value and TotalPopulation
health_df['TotalPatients'] = (
    health_df['Data_Value'] * health_df['TotalPopulation'] / 100
).astype('int64')  # Convert the result to 'int64' type

# Calculate the total number of patients aged 18 and above based on the Data_Value and TotalPop18plus
health_df['TotalPatients18plus'] = (
    health_df['Data_Value'] * health_df['TotalPop18plus'] / 100
).astype('int64')  # Convert the result to 'int64' type
```

#### Disability Analysis in Michigan's Top 5 Most Populated Counties

Let‚Äôs identify the top 5 counties in Michigan with the highest total population by grouping the data by **LocationName** and summing the **TotalPopulation**. We will then filter the dataset to include only data related to **Disability** (excluding the **Any Disability** category, as that's the sum of all disabilities) and focus on the selected counties. 

These two aggregates will be visualized in two **bar charts**:
- The first chart will display the **percentage** of people affected by different types of disabilities in the top 5 most populated counties.
- The second chart will show the **total number** of people affected by each disability type in these counties.

We are doing this step to determine whether percentage or the total patient population is better for visualization purposes. 


```python
# Identify the top 5 counties with the largest total population
top_counties = michigan_df.groupby("LocationName")["TotalPopulation"].sum().nlargest(5).index

# Filter the DataFrame for data related to the "Disability" category
top_county_data = michigan_df[
    (michigan_df["Category"] == "Disability") &  #Filter for Disability category
    (michigan_df["Short_Question_Text"] != "Any Disability") &  # Exclude "Any Disability"
    (michigan_df["LocationName"].isin(top_counties))  #Filter for top 5 counties
]

# Group the filtered data by 'Short_Question_Text' and 'LocationName', 
# sum 'Data_Value' (percentage), and unstack the data to create a matrix format
disability_percentage = (
    top_county_data.groupby(["Short_Question_Text", "LocationName"])["Data_Value"]
    .sum()
    .unstack()  # Convert 'LocationName' to columns
    .fillna(0)  # Replace missing values with 0
)

# Similarly, group by 'Short_Question_Text' and 'LocationName', 
# sum 'TotalPatients' (total people affected), and unstack the data
disability_total = (
    top_county_data.groupby(["Short_Question_Text", "LocationName"])["TotalPatients"]
    .sum()
    .unstack()  # Convert 'LocationName' to columns
    .fillna(0)  #replace missing values with 0
)

# Create a 1x2 grid of subplots for visualizing the data
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# First plot: Bar plot showing the percentage of people affected by different disabilities
disability_percentage.plot(kind="barh", colormap="viridis", ax=axes[0])
axes[0].set_xlabel("Percentage Affected")  # Label for the x-axis
axes[0].set_ylabel("Disability Type")  # Label for the y-axis
axes[0].set_title("Disability Type by Percent Affected in Top 5 Populated Counties")  # Title for the plot

# Second plot: Bar plot showing the total number of people affected by different disabilities
disability_total.plot(kind="barh", colormap="viridis", ax=axes[1])
axes[1].set_xlabel("Total People Affected")  # Label for the x-axis
axes[1].set_ylabel("Disability Type")  # Label for the y-axis
axes[1].set_title("Disability Type by Total People Affected in Top 5 Populated Counties")  # title for the plot

# Adjust the layout to prevent overlap of titles and labels
plt.tight_layout()

# Display the plots
plt.show()
```

We can infer the following:

- **Percentage of People Affected**:
The first chart, showing the percentage of the population affected by different disabilities, reveals that **mobility disabilities** and **cognitive disabilities** are the most common conditions, with significant proportions of the population in these counties affected. Wayne County has the highest percentage, while Oakland County has the lowest percentage of individuals affected.

- **Total Number of People Affected**:
The second chart, displaying the total number of people affected by disabilities, highlights counties with larger populations, which somewhat skews the results. For example, although **mobility disabilities** affect a significant percentage of people, this chart doesn't account for population size. Larger counties with higher populations, like Wayne County, show significantly higher numbers of people affected, while other counties with smaller populations show lower totals. This can potentially mislead our understanding of the true disability burden across the state.

Therefore, percentage chart is a more **reliable and insightful** method for analyzing the prevalence of disabilities, especially when comparing counties of varying sizes. The total number of affected individuals can be misleading in areas with larger populations, making percentages a more accurate way to understand how widespread each disability is across different regions in Michigan. We can make this visualization even clearer to read using a **100% stacked bar chart**.

```python
# Create a 3x2 grid of subplots for visualizing different health categories
fig, axes = plt.subplots(3, 2, figsize=(20, 12))  # Define figure size
fig.suptitle("Health Data Analysis by Category in Top 5 Populated Counties (100% Stacked)", fontsize=16)  
# Set the overall title for the figure

# Flatten the 3x2 axes array into a 1D array for easier iteration
axes = axes.flatten()

# Loop through each category in the predefined list and create 100% stacked bar charts
for i, category in enumerate(categories):
    # Filter the dataset to include only data for the selected category and top 5 populated counties
    category_data = michigan_df[
        (michigan_df["Category"] == category) &  # Select specific health category
        (michigan_df["LocationName"].isin(top_counties))  # Restrict to top 5 counties
    ]

    # Group by 'Short_Question_Text' and 'LocationName', summing up 'Data_Value' (percentage data)
    category_grouped = (
        category_data.groupby(["Short_Question_Text", "LocationName"])["Data_Value"]
        .sum()
        .unstack()  # Convert 'LocationName' values into columns
        .fillna(0)  # Replace NaN values with 0
    )

    # Convert data to percentages to create a 100% stacked bar chart
    category_grouped = category_grouped.div(category_grouped.sum(axis=1), axis=0) * 100  

    # Plot the 100% stacked horizontal bar chart
    category_grouped.plot(kind="barh", stacked=True, colormap="icefire", ax=axes[i])

    # Set axis labels and titles
    axes[i].set_xlabel("Percentage Distribution")  # X-axis label
    axes[i].set_ylabel("Measure")  # Y-axis label
    axes[i].set_title(f"{category} by Percent Affected in Top 5 Counties (100% Stacked)")  # Set title

# Adjust the layout to prevent overlap of elements
plt.tight_layout(rect=[0, 0.1, 1, 0.98])

# Display the figure
plt.show()
```

### Census ACS Poverty/ Income Data (2022) Merge

We are now going to pull poverty-related data from the **U.S. Census Bureau's ACS API** for all counties in the 2022 ACS 5-Year dataset. 

```python
# Load the CSV file
mapping = pd.read_csv(os.path.join(data_dir, "acs_poverty_md.csv"))  

# Extract the column names for the API request
columns = mapping["Column Name"].tolist()

# Construct the API URL
base_url = "https://api.census.gov/data/2022/acs/acs5/subject"
params = ",".join(columns)
query_url = f"{base_url}?get={params}&for=county:*"

# Call the API
response = requests.get(query_url)
if response.status_code != 200:
    print(f"Error: {response.status_code}, {response.text}")
else:  
    data = response.json()
    
    # Convert response to DataFrame
    census_df = pd.DataFrame(data[1:], columns=data[0])
    
    # Create a mapping from "Column Name" to "Label"
    column_mapping = dict(zip(mapping["Column Name"], mapping["Label"]))
    
    # Rename columns in census_df
    census_df.rename(columns=column_mapping, inplace=True)
    
    # Replace -999999999 values with 0
    census_df.replace("-999999999", "0", inplace=True)
    census_df.replace("-888888888", "0", inplace=True)
    census_df.replace("-666666666.0", "0", inplace=True)
    
    # Save to CSV
    census_df.to_csv(os.path.join(data_dir, "poverty_census_data.csv"), index=False)
    print("Data saved as poverty_census_data.csv")
```

```python
# Split 'Geographic Area Name' into State and County
census_df[['County', 'State']] = census_df['Geographic Area Name'].str.split(',', expand=True)

# Strip extra spaces
census_df['County'] = census_df['County'].str.strip().str.split().str[0]
census_df['State'] = census_df['State'].str.strip()

#Display the result
census_df.head()
```

We can see that all the numerical columns are stored as objects in the Census ACS dataset. Therefore, we need to convert these columns from object to `int64` or `float64` for proper numerical analysis.

```python
# Extract the list of column labels from the 'Label' column in the mapping DataFrame.
label = mapping["Label"].tolist()

# Iterate through each column name in the extracted label list.
for column in label:
    # Check if the column is one of the specified columns that should be converted to float64.
    if column in ['Households<$10k', 'Households>$200k', 'BP_ Total%', 'BP_ African American%', 'BP_ American Indian/ Alaska Native%', 'BP_ Asian%', 'BP_ Hawaiian/ Pacific Islander%', 'BP_ Two or more races%', 'BP_ Other race%', 'BP_ Hispanic or Latino%', 'BP_ White%']:
        census_df[column] = census_df[column].astype('float64')  # Convert column to float64 type.
    
    # If the column is not 'Geography' or 'Geographic Area Name', convert it to int64.
    elif column not in ['Geography', 'Geographic Area Name']:
        census_df[column] = census_df[column].astype('int64')  # Convert column to int64 type.
```

Finally, let's display **health data by category** for both the **highest and lowest poverty counties**, which will allow us to compare the health metrics across various categories between high and low poverty areas and possibly give us some more insights into the health disparities between the two extreams.

```python
def plot_health_data_by_category(df):
        # Define the list of categories to filter the data by
    categories = [
        'Health Outcomes', 'Disability', 'Health Status', 'Prevention', 
        'Health-Related Social Needs', 'Health Risk Behaviors'
    ]  # Select categories for analysis

    # Create a 2x3 grid of subplots to accommodate the six categories
    fig, axes = plt.subplots(3, 2, figsize=(18, 12))
    fig.suptitle("Health Data Analysis by Category", fontsize=16)

    # Flatten the axes array to make iteration easier
    axes = axes.flatten()

    # Loop through each category to create a bar chart for each
    for i, category in enumerate(categories):
        # Filter data for the current category, group by 'Health_Measure', 
        # sum 'Pop_Percent', and sort values in descending order
        category_data = (
            df[df['Category'] == category]
            .groupby('Health_Measure')['Pop_Percent']
            .sum()
            .sort_values(ascending=False)
        )
        
        # Create a bar plot for the current category
        sns.barplot(x=category_data.values, y=category_data.index, palette="crest", ax=axes[i])
        
        # Set the x-axis label and format the x-axis tick labels to prevent scientific notation
        axes[i].set_xlabel("Percent of People Affected")
        axes[i].ticklabel_format(style='plain', axis='x')
        
        # Set the y-axis label and title for the subplot
        axes[i].set_ylabel(category)
        axes[i].set_title(f"{category} by Percent of People Affected")
        
    # Adjust the layout to prevent overlap of titles and labels
    plt.tight_layout(rect=[0, 0.1, 1, 0.98])

    # Display the plots
    plt.show()
```

```python
plot_health_data_by_category(highest_poverty_df) # for Isabella
```

```python
plot_health_data_by_category(lowest_poverty_df) # for Livingston
```

We can see that in **Isabella County**, which has the highest poverty rate, the general health of the population appears to be poorer. People in this county are more likely to rely on health-related social services and face challenges such as food insecurity. Additionally, there is a greater prevalence of health risk behaviors in this county, likely as a result of the increased stress and economic hardships associated with higher poverty levels.

On the other hand, **Livingston County**, which has the lowest poverty rate, presents a different health challenge. Despite having fewer people living in poverty, the county shows a higher incidence of binge drinking, which is a significant health risk behavior.
